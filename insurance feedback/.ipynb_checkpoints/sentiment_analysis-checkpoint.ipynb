# SENTIMENTAL ANALYSIS 

```python
import pandas as pd
from textblob import TextBlob
```

```python
#load data set
df=pd.read_excel("Insurance+Customer+Feedback.xlsx")
```

```python
df.head(5)
```

```python
# function for sentiment polarity
def sentiment_polarity(text):
    if pd.isna(text):
         return 0
    return TextBlob(text).sentiment.polarity
```

```python
# add sentiment score
df['score sentiment']=df['Feedback'].apply(sentiment_polarity)
df.head(5)
```

```python
# sentiment label
df['label']=df['score sentiment'].apply(lambda x: 'positive' if x>0 else 'negative' if x<0 else 'neutral')
df.head(5)
```

```python
# visualisation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
count_customer_name=df.groupby('label')['Customer Name'].count()
count_customer_name.plot(kind='barh',color='teal')
plt.ylabel('count of customers')
plt.xlabel('feedback')
plt.title('count of customer by feedback')
plt.show()
```
